# K-matrix

Let's study the K-matrix in this notebook. The K-matrix is a matrix that describes the interaction between particles in a scattering process. It is defined as the inverse of the T-matrix, which describes the transition amplitude between initial and final states.

*the description below is taken from BAD 01237v004*

In the K-matrix formalism transition amplitudes are defined as
$$
F_l = \sum_j \left(I - i K \rho\right)^{-1}_{lj} P_j
$$
where $I$ is the identity matrix, $K$ is the matrix describing the S-wave scattering process, $\rho$ is the phase-space matrix and $P$ is the initial production vector.

The index $l$ is the pion channel index, where 1=ππ, 2=KK, 3=multi−meson, 4=ηη, and 5=ηη′.

In this picture, the production process can be viewed as consisiting of an initial preparation of several states, which are then propagated by the $\left(I - i K \rho\right)^{-1}_{lj}$ term into the final state. Since we are describing the ππ channel, only the $F_1$ amplitude is present. See the figure below for a schematic representation of the K-matrix formalism.

![image](img/kmatrix/intuitive_kmatrix.png)

$P_j$ is the production vector,
$$
P_j(s) = \left\{\sum_{\alpha}\frac{\beta_{\alpha}g_j^{(\alpha)}}{m_{\alpha}^2-s} + f_{1j}^{\text{prod}}\frac{1.0 -s_0^{\text{prod}}}{s -s_0^{\text{prod}}}\right\}.
$$
The free parameters for the P-vector are $\beta_{\alpha}$, $f_{1j}^{\text{prod}}$ and $s_0^{\text{prod}}$, which are determined in the fit.

The masses and widths of the Breit-Wigner components are mostly taken from PDG. However, there are no corresponding K-matrix parameters listed in the PDG. We therefore obtain these parameters from the study of Anisovich and Sarantsev (AS) [7], which performed a global fit of the available ππ scattering data from threshold up to 1900 MeV/c2. The AS K-matrix parameterization is
$$
K_{ij}(s) = \left\{\sum_{\alpha}\frac{g_i^{(\alpha)}g_j^{(\alpha)}}{m_{\alpha}^2-s} + f_{ij}^{\text{scat}}\frac{1.0 -s_0^{\text{scat}}}{s -s_0^{\text{scat}}}\right\} \frac{1-s_{A0}}{s-s_{A0}}(s-s_Am^2_{\pi}/2)
$$
where $g_i^{(\alpha)}$ is the coupling constant of the K-matrix pole $m_{\alpha}$ to the $i^{\text{th}}$ channel. The parameters $f_{ij}^{\text{scat}}$ and $s_0^{\text{scat}}$ describe the slowly-varying part of the K-matrix element. The factor $\frac{1-s_{A0}}{s-s_{A0}}(s-s_Am^2_{\pi}/2)$ is the Adler zero term. The parameter values used in this analysis are listed in Table 8 (obtained through private communication with AS).

By definition, $K$ is real and symmetric. However, the K-matrix coupling and pole do not have a physical meaning. The K-matrix is related to the physical (and therefore, observable) T-matrix by the following expression:
$$
T = \left(I - i K \rho\right)^{-1} K
$$
A one-to-one correspondance between $K$ and $T$ exists only in the simplest case of a single pole and a single channel.

The phase space matrix is diagonal: $\rho_{ab} = \delta_{ab}\rho_a$ where
$$
\rho_i(s) = \sqrt{1- \frac{(m_{1i}+m_{2i})^2}{s}}
$$
The normalization is such that $\rho_i\to 1$ as $s\to\infty$. Since S-matrix has to respect analyticity, we
used an analytic continuation for $\rho_a$ below the threshold. The expression of the multi-meson state
phase space is written as
$$
\rho_5(s) = \begin{cases}
                \rho_{51}\qquad s < 1 GeV \\
                \rho_{52}\qquad s > 1 GeV 
                \end{cases},
$$
where
$$
\begin{align*}
\rho_{51}(s) &= \rho_0\int \int \frac{ds_1}{\pi}\frac{ds_2}{\pi}\frac{M^2_{\rho}\Gamma(s_1)\Gamma(s_2)\sqrt{(s+s_1-s_2)^2-4ss_1}}{s\left[(M^2-s_1^2)^2+M^2\Gamma^2(s_1)\right]\left[(M^2-s_2^2)^2+M^2\Gamma^2(s_2)\right]}\\
\rho_{52}(s) &= \left(\frac{s-16m^2_{\pi}}{s}\right)
\end{align*}
$$
Here $s_1$ and $s_2$ are the squared energies of the two pions, $M$ is the $\rho$ meson mass and $\Gamma(s)$ is the energy-dependent width. The factor $\rho_0$ provides the continuity of $\rho_5(s)$ at $s = 1GeV^2$. Energy conservation in the dipion system must be satified when calculating the integral. This complicated
expression reveals the fact that the $\rho$ meson has an intrinstic width. If one sets $\Gamma(s) = \delta(s)$, where $\delta$ is the Dirac $\delta$ function, the usual twobody phase space factor is obtained.

### Production vector
The production vector $P_j$ is defined as
$$
P_j(s) = \left\{\sum_{\alpha}\frac{\beta_{\alpha}g_j^{(\alpha)}}{m_{\alpha}^2-s} + f_{1j}^{\text{prod}}\frac{1.0 -s_0^{\text{prod}}}{s -s_0^{\text{prod}}}\right\}.
$$

```python
def prod_vector_nr(s, f1, s0):
    nr = f1*(1.0-s0)/(s-s0)
    return nr

def prod_vector(s, b, ma, g, f1, s0):
    nr = f1*(1.0-s0)/(s-s0)
    res = complex(0.0, 0.0)
    for a in range(len(ma)):
        res += (b[a] * g[a] / (ma[a]**2 - s))
    return res + nr
```

Define the K-matrix
$$
K_{ij}(s) = \left\{\sum_{\alpha}\frac{g_i^{(\alpha)}g_j^{(\alpha)}}{m_{\alpha}^2-s} + f_{ij}^{\text{scat}}\frac{1.0 -s_0^{\text{scat}}}{s -s_0^{\text{scat}}}\right\} \frac{1-s_{A0}}{s-s_{A0}}(s-s_Am^2_{\pi}/2)
$$

```python
def Kmatrix_with_parameters(s, g_ia, m_a, fscat_ij, s0scat, sA0, sA):
    ndim = int(len(g_ia))
    ndim_a = int(len(m_a))
    K = np.zeros((ndim, ndim))
    nr_factor = (1.0 - s0scat) / (s - s0scat)
    for i in range(ndim):
        for j in range(ndim):
            K[i][j] += fscat_ij[i][j] * nr_factor
            for a in range(ndim_a):
                K[i][j] += (g_ia[i][a] * g_ia[j][a]) / (m_a[a]**2 - s)
    K *= (1-sA0) / (s - sA0) * (s - 0.5*sA*mpi**2)
    return K

def Kmatrix(s):
    return Kmatrix_with_parameters(s, g_ia, m_a, fscat_ij, s0scat, sA0, sA)
```

Let's now calculate the transition amplitude
$$
F_l = \sum_j \left(I - i K \rho\right)^{-1}_{lj} P_j
$$
Start from the D-matrix
$$
D = \left(I - i K \rho\right)^{-1}
$$

```python
def get_rho(s):
    rho = np.zeros((5,5),dtype=complex)
    ps = np.array([ two_body_ps(s, mpi, mpi),
                    two_body_ps(s, mkz, mkz),
                    multi_meson_ps(s, mpi),
                    two_body_ps(s, meta, meta),
                    two_body_ps(s, meta, metap) ] )
    for i, v in enumerate(ps):
        rho[i][i] = v
    return rho
rho_ij = np.array( [ get_rho(s) for s in s_pl])
```

```python
def D(K, rho):
    I = np.identity(5)
    invD = I - 1j*np.matmul(K, rho)
    D = np.linalg.inv(invD)
    return D
```

```python
D_ij = np.array( [ D(K_ij[i], rho_ij[i]) for i in range(len(s_pl))])
```

```python
fig, ax = plt.subplots(5,5, figsize=(15,15))
for i in range(5):
    for j in range(5):
        ax[i][j].plot(s_pl, np.real(D_ij[:,i,j]), label='real')
        ax[i][j].plot(s_pl, np.imag(D_ij[:,i,j]), label='imaginary')
        ax[i][j].plot(s_pl, density(D_ij[:,i,j]), label='density')
        ax[i][j].set_xlabel(r's ($GeV^2$)')
        ax[i][j].set_ylabel(r'$D_{'+str(i+1)+str(j+1)+'}$')
        ax[i][j].set_yscale('log')
ax[0][0].legend()
plt.tight_layout()
```

## K-matrix in a single function

Let's write a function that calculates the K-matrix for the ππ channel for a set of given parameters.

```python
def KmatrixAmplitude(s, g_ia, m_a, fscat_ij, s0scat, sA0, sA, s0prod, beta, fprodj):
    Kij = Kmatrix_with_parameters(s, g_ia, m_a, fscat_ij, s0scat, sA0, sA)
    rho = get_rho(s)
    Dij = D(Kij, rho)
    pj = np.array( [prod_vector(s, beta, m_a, g_ia[i], fpj, s0prod) for i, fpj in enumerate(fprodj)])
    F1 = Fl(Dij, pj)[0]
    return F1
```

```python
def plot_amplitude(amp):
    fig, ax = plt.subplots(1,2, figsize=(12,8))
    ax[0].plot(s_pl, density(amp))
    ax[0].set_xlabel(r's ($GeV^2$)')
    ax[0].set_ylabel('Density')
    ax[1].plot(np.real(amp), np.imag(amp), 'o')
    ax[1].set_xlabel('Real part')
    ax[1].set_ylabel('Imaginary part')
    ax[1].set_title('Argand plot')
    yabs_max = abs(max(ax[1].get_ylim(), key=abs))
    ax[1].set_ylim(ymin=-yabs_max, ymax=yabs_max)
    xabs_max = abs(max(ax[1].get_xlim(), key=abs))
    ax[1].set_xlim(xmin=-xabs_max, xmax=xabs_max)
    ax[1].grid()
    return fig, ax
```

```python
km = [KmatrixAmplitude(s, g_ia, m_a, fscat_ij, s0scat, sA0, sA, s0prod, beta, fprodj) for s in s_pl]
```

```python
fig, ax = plot_amplitude(km)
```

## TensorFlowAnalysis Implementation
Let's implement the K-matrix in TensorFlowAnalysis. One needs
- phase-space factors
- production vector
- K-matrix
- transition amplitude
- K-matrix in a single function

Let's start with the phase-space factors

```python
# Import tensorflow
import tensorflow as tf

# Import AmpliTF modules
import amplitf.interface as atfi
```

```python
@atfi.function
def two_body_phsp_tf(s, ma, mb):
    km_phsp = atfi.sqrt( atfi.cast_complex((atfi.const(1.0) - atfi.pow(ma+mb,2)/s)) )
    return km_phsp

@atfi.function
def multi_meson_phsp_tf(s, m):
    km_phsp = tf.where(s > 1.0, atfi.sqrt(atfi.cast_complex(atfi.const(1.0) - 16.0*m**2/s)), 
                                atfi.cast_complex(0.0005- 0.0193*s + 0.1385*s**2 - 0.2084*s**3 - \
                                0.2974*s**4 + 0.1366*s**5 + 1.0789*s**6) )
    return km_phsp

@atfi.function
def KM_phsp(s, masses):
    km_phsp = tf.stack([
        two_body_phsp_tf(s, masses[0][0], masses[0][1]),
        two_body_phsp_tf(s, masses[1][0], masses[1][1]),
        multi_meson_phsp_tf(s, masses[2][0]),
        two_body_phsp_tf(s, masses[3][0], masses[3][1]),
        two_body_phsp_tf(s, masses[4][0], masses[4][1])
    ], axis=1)
    return km_phsp
```

```python
masses_poles = [[atfi.const(mpi), atfi.const(mpi)], 
                [atfi.const(mkz), atfi.const(mkz)], 
                [atfi.const(mpi)], 
                [atfi.const(meta), atfi.const(meta)], 
                [atfi.const(meta), atfi.const(metap)]]
km_phsp_tf = KM_phsp(s_pl, masses_poles)
```

```python
def prod_vector_nr_tf(s, f1, s0):
    s_exp = tf.expand_dims(s, axis=1)
    f1_exp = tf.expand_dims(f1, axis=0)
    #nr = f1_exp*(atfi.const(1.0)-s0)/(s_exp-s0)
    nr = f1_exp*tf.cast((atfi.const(1.0)-s0)/(s_exp-s0), dtype=atfi.ctype())
    return nr

def prod_vector_res_tf(s, b, ma, g):
    b_exp = tf.expand_dims(b, axis=0) # [1, n]
    ma_exp = tf.expand_dims(ma, axis=0) # [1, n]
    s_exp = tf.expand_dims(s, axis=1) # [k, 1]
    ta = b_exp / (ma_exp**2 - s_exp) # [k, n]
    ta_exp = tf.expand_dims(ta, axis=1) # [k, 1, n]
    g_exp = tf.expand_dims(g, axis=0) # [1, n, n]
    res = tf.reduce_sum(ta_exp * g_exp, axis=2)
    return res

def prod_vector_tf(s, b, ma, g, f1, s0):
    # f1_exp = tf.expand_dims(f1, axis=0)
    # ma_exp = tf.expand_dims(ma, axis=0)
    # b_exp = tf.expand_dims(b, axis=0)
    # s_exp = tf.expand_dims(s, axis=1)
    # g_exp = tf.expand_dims(g, axis=0)
    #nr = f1_exp*tf.cast((atfi.const(1.0)-s0)/(s_exp-s0), dtype=atfi.ctype())
    nr = prod_vector_nr_tf(s, f1, s0)
    res = prod_vector_res_tf(s, b, ma, g)
    #res = tf.reduce_sum(b_exp * g_exp / (ma_exp**2 - s_exp), axis=2)
    return res + nr
```
